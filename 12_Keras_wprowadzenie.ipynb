```python
from tensorflow.keras.models import Sequential

model = Sequential()
print(model)
```

```python
from tensorflow.keras.layers import Dense
```

```python
model.add(Dense(units=4, input_shape=(10, )))  # units - liczba neuronów, input_shape - rozmiar danych wejściowych - definiuje liczbę neuronów wejsciowych
```

model.summary()

```python
model.add(Dense(units=2))  # dodajemy drugą warstwę ukrytą
model.summary()            # 10 parametrów  ==> (4 + 1) * 2
```

# Podsumowanie

```python
model = Sequential()
model.add(Dense(units=8, activation='relu', input_shape=(10, )))
model.add(Dense(units=4, activation='sigmoid'))  # warstwa końcowa ilośc wyniików == 4
model.summary()
```

```python
model = Sequential()
model.add(Dense(units=8, activation='relu', input_shape=(10, )))
model.add(Dense(units=1, activation='sigmoid'))  # warstwa końcowa ilośc wyniików == 1
model.summary()
```

# Kompilacja modelu - przypisanie optymalizatora (optimizers), funkcji straty(loss functions) oraz metryki(metrics).

```python
# klasyfikacja binarna
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# klasyfikacja wieloklasowa
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# regresja
model.compile(optimizer='rmsprop',
              loss='mse')
```
