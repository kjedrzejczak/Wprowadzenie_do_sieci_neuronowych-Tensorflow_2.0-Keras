# Przykad klasyfikacji wieloklasowej

```python
data = np.random.random((1000, 150))
labels = np.random.randint(10, size=(1000, 1))
```

```python
np.unique(labels)
```

```python
print(data.shape)
print(labels.shape)
```

```python
from tensorflow.keras.utils import to_categorical  # kodowanie 0 1

labels = to_categorical(labels, num_classes=10)
labels[1]
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(150, )))
model.add(Dense(units=10, activation='softmax'))  # 10 kategorji wic 10 neuron贸w na wyjciu

model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',     # categorical 垛锔
              metrics=['accuracy'])

history = model.fit(data, labels, epochs=30, batch_size=32, validation_split=0.2, verbose=1)
```

```python
test_data = np.random.random((10, 150))

model.predict(test_data)
```

```python
f = model.predict(test_data)
np.where(f[1] == max(f[1]))   #  dla pierwszej pr贸bki
```

# Przykad regresji

```python
data = np.random.random((1000, 150))
labels = 50 * np.random.random(1000)
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(150, )))
model.add(Dense(units=1))  # jedna warto

model.compile(optimizer='rmsprop',
              loss='mse')

model.fit(data, labels, epochs=30, batch_size=32, validation_split=0.2, verbose=1)
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_shape=(150, )))
model.add(Dense(units=1))  # jedna warto

model.compile(optimizer='rmsprop',
              loss='mae',
              metrics=['mse'])

model.fit(data, labels, epochs=30, batch_size=32, validation_split=0.2, verbose=1)
```

```python
test_data = np.random.random((10, 150))

model.predict(test_data)
```
