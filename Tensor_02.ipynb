Wektor

```python
import numpy as np

vector = np.array([2, 4, -6, 5])
print(vector)
print(type(vector))
print(vector.shape)
print(vector.dtype)
print(vector.ndim)
print(len(vector))
```

```python
vector = np.array([[2, 4, -6, 5], dtype='float')
print(vector)
print(type(vector))
print(vector.shape)
print(vector.dtype)
print(vector.ndim)
print(len(vector))
```

```python
vector = np.array([[2, 4, -6, 5], [1, 2, 3, 4]], dtype='float')
print(vector)
print(type(vector))
print(vector.shape)
print(vector.dtype)
print(vector.ndim)
print(len(vector))
```

Tensor

```python
tensor = np.array([[[2, 4, -6, 5],
                    [1, 2, 3, 4]],
                  [[2, 4, -6, 5],
                   [1, 2, 3, 4]]], dtype='float')
print(tensor)
print(type(tensor))
print(tensor.shape)
print(tensor.dtype)
print(tensor.ndim)
print(len(tensor))
```

```python
tensor = np.array([[[2, 4, -6, 5],
                    [1, 2, 3, 4]],
                   [[2, 4, -6, 5],
                   [1, 2, 3, 4]],
                   [[2, 4, -6, 5],
                    [1, 2, 3, 4]]], dtype='float')
print(tensor)
print(type(tensor))
print(tensor.shape)
print(tensor.dtype)
print(tensor.ndim)
print(len(tensor))
```
